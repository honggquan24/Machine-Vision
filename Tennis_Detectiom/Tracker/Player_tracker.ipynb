cls (class): Loại: tensor
Mô tả: Chứa các ID lớp của các đối tượng được phát hiện. Mỗi số tương ứng với một danh mục cụ thể mà mô hình đã được huấn luyện để nhận dạng (ví dụ, 0 cho 'người', 1 cho 'xe đạp', v.v.).

conf (độ tin cậy):Loại: tensor
Mô tả: Chứa các điểm tin cậy cho mỗi đối tượng được phát hiện. Các giá trị này chỉ độ tin cậy của mô hình trong việc phát hiện, thường nằm trong khoảng từ 0 đến 1, trong đó các giá trị cao hơn đại diện cho độ tin cậy cao hơn.

data: Loại: tensor
Mô tả: Chứa tọa độ hộp giới hạn chi tiết và thông tin khác cho mỗi phát hiện theo định dạng [x1, y1, x2, y2, cls, conf, id]:
x1, y1: Góc trên bên trái của hộp giới hạn.
x2, y2: Góc dưới bên phải của hộp giới hạn.

cls: ID lớp của đối tượng được phát hiện.
conf: Điểm tin cậy.
id: Định danh duy nhất cho đối tượng được theo dõi (hữu ích trong việc theo dõi video).
id: Loại: tensor
Mô tả: Chứa các định danh duy nhất cho mỗi đối tượng được phát hiện và theo dõi. Các ID này giúp theo dõi đối tượng qua các khung hình khác nhau trong video.

is_track: Loại: bool
Mô tả: Chỉ ra liệu việc phát hiện có đang được theo dõi qua các khung hình hay không. Được đặt thành True nếu việc theo dõi được kích hoạt.

orig_shape:Loại: tuple
Mô tả: Kích thước ban đầu của hình ảnh đầu vào dưới dạng (chiều cao, chiều rộng).

shape: Loại: torch.Size
Mô tả: Hình dạng của tensor chứa dữ liệu phát hiện, chỉ số lượng phát hiện và các điểm dữ liệu cho mỗi phát hiện.

xywh:Loại: tensor
Mô tả: Tọa độ hộp giới hạn dưới dạng [x_center, y_center, width, height] cho mỗi phát hiện:
x_center, y_center: Tọa độ trung tâm của hộp giới hạn.
width, height: Kích thước của hộp giới hạn.

xywhn:Loại: tensor
Mô tả: Tọa độ hộp giới hạn được chuẩn hóa dưới dạng [x_center, y_center, width, height] tương đối với kích thước của hình ảnh. Các giá trị được chuẩn hóa để nằm trong khoảng từ 0 đến 1:
x_center, y_center: Tọa độ trung tâm chuẩn hóa của hộp giới hạn.
width, height: Kích thước chuẩn hóa của hộp giới hạn.

```python
class PlayerTracker():
    def __init__(self, model_path):
        self.model = YOLO(model_path)

    def dict_bounding_box(self, frame):
        # object of box
        result_box = self.model.track(frame, persist= True) [0]
        # name_class 
        name_class = result_box.names
        # x1, y1, x2, y2 
        bounding_box_dict = {}

        for box in result_box.boxes:
            id = int((box.id.tolist()[0]))
            inf_box =  box.xyxy.tolist()[0]
            id_of_class = int(box.cls.tolist()[0])
            name_of_class = name_class[id_of_class]
            if name_of_class == 'person':
                bounding_box_dict[id] = inf_box
        
        return bounding_box_dict
    def detect_boundingbox_frames(self, frames, read_from_stubs= False, stub_path= None):
        player_bounding_box_detection = []
        
        # if read_from_stubs and stub_path is not None:
        #     try:
        #         with open(stub_path, 'rb') as f:
        #             player_bounfing_box_detection = pickle.load(f)
        #     except FileNotFoundError:
        #         print(f"Warning: File not found at {stub_path}, proceeding with fresh detection.")
        #     except Exception as e:
        #         print(f"Error loading detection stubs: {e}")
        #     else:
        #         return player_detection

        for frame in frames:
            bounding_box = self.dict_bounding_box(frame)
            player_bounding_box_detection.append(bounding_box)

        # if stub_path is not None:
        #     try:
        #         with open(stub_path, 'wb') as f:
        #             pickle.dump(player_detection, f)
        #     except Exception as e:
        #         print(f"Error saving detection stubs to {stub_path}: {e}")
        return player_bounding_box_detection

    def draw_bounding_box(self, video_frames, player_bounding_box_detection):
        outputvideoframe = []
        for frame, bounding_box_dict in zip(video_frames, player_bounding_box_detection):
            for id, bounding_box in bounding_box_dict.items():
                x1, y1, x2, y2 = bounding_box
                cv.rectangle(img= frame, 
                              pt1= (int(x1), int(y1)), 
                              pt2= (int(x2), int(y2)),
                              color= (255, 255, 255),
                              thickness= 1)
                cv.putText(img= frame,
                            text= f'Player {id}', 
                            org= (int(x1), int(y1)),
                            fontFace= cv.FONT_HERSHEY_COMPLEX,
                            fontScale= 1,
                            color= (122, 122, 122),
                            thickness= 1)
            outputvideoframe.append(frame)
        return outputvideoframe
```

```python
def read_video(video_path):
    cap = cv.VideoCapture(video_path)
    frames = []
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frames.append(frame)
    cap.release()
    return frames
```

```python
import cv2
import os

def extract_frames(video_path, output_dir, frame_rate=1):
    # Create output directory if it doesn't exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    # Open the video file
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print("Error: Could not open video.")
        return
    
    frame_count = 0
    saved_frame_count = 0
    
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Save every 'frame_rate' frame
        if frame_count % frame_rate == 0:
            output_path = os.path.join(output_dir, f"frame_{saved_frame_count:04d}.png")
            cv2.imwrite(output_path, frame)
            saved_frame_count += 1
        
        frame_count += 1
    
    cap.release()
    print(f"Extracted and saved {saved_frame_count} frames to {output_dir}")

# Example usage
video_path = r"C:\Users\ADMIN\Downloads\Pothole Videos\Pothole Videos\pothole_video\pothole_video\test\rgb\0081.mp4"
output_dir = r"C:\Users\ADMIN\Downloads\Test"
extract_frames(video_path, output_dir, frame_rate=1)
```

```python
import os

def rename_files(directory, prefix='image'):
    # List all files in the directory
    files = sorted(os.listdir(directory))
    
    for i, filename in enumerate(files):
        # Construct new file name
        new_name = f"{prefix}_{i + 1}.png"
        
        # Get full file paths
        old_file = os.path.join(directory, filename)
        new_file = os.path.join(directory, new_name)
        
        # Rename the file
        os.rename(old_file, new_file)
        print(f"Renamed '{filename}' to '{new_name}'")

# Example usage
directory = r"C:\\Data\\Machine Vision\\Tennis_Detectiom\\Output"
rename_files(directory)
```

```python
import os

def delete_files(directory):
    # List all files in the directory
    files = os.listdir(directory)
    
    for i, filename in enumerate(files):
        file_path = os.path.join(directory, filename)
        
        # Check if it is a file and not a directory
        if i % 2 == 0:
            os.remove(file_path)

# Example usage
directory = r"C:\\Data\\Machine Vision\\Tennis_Detectiom\\Output"
delete_files(directory)
```
